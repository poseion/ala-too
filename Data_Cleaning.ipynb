### Discuss the importance of feature scaling in machine learning.
* Feature scaling ensures that all features are on a similar scale. It’s essential for many machine learning algorithms
### Explain the difference between min-max scaling and z-score normalization.
* Min-max scaling (normalization) rescales features to a range between 0 and 1.
Z-score normalization (standardization) transforms features to have a mean of 0 and a standard deviation of 1.
